```python
def L2_dist(v1, v2):
    ret = 0
    #print (type(v1))
    for i in range(v1.shape[0]):
        ret = ret + (v1[i] - v2[i])**2
    return ret
```

```python
def L2_dist_calc(vector_list1, vector_list2):
    L2_diff = [L2_dist(vector_list1[id], vector_list2[id]) for id in range(len(vector_list1))]
    return L2_diff
```
